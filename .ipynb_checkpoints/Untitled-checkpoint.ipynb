```python

import numpy as np
import pandas as pd 
import os 
base_url = os.path.expanduser('/root/Downloads/jupyter/data/') 
# create a dict of strings with pd name as key and filename as value 
files_dict = {} 
files = os.listdir(base_url) 
for f in files: 
    if f.endswith('.tsv'): 
        key = f[:-len('.tsv')] 
        key = key.replace(" ", "_") 
        key = key.replace("-", "_") 
        files_dict[key] = f 
        for k, v in files_dict.items():
            exec("%s = pd.read_csv(base_url + '%s',sep='\t')" % (k, v), globals()) 
```

```python
globals()
```

```python
#create all_dfs dicts to hold DataFrames 
all_dfs = files_dict #add DFs 
for k, v in all_dfs.items():
    exec("all_dfs['%s'] = %s" % (k, k))
```
